```python
# Importing libraries

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from lab_utils_multi import load_house_data
```

```python
# load dataset
X_train, y_train = load_house_data()
X_features = ['size(sqft)', 'bedrooms', 'floors', 'age']
```

```python
scaler = StandardScaler()
X_norm = scaler.fit_transform(X_train)
```

```python
sgd = SGDRegressor(max_iter=1000)
sgd.fit(X_norm, y_train)

w_norm = sgd.coef_
b_norm = sgd.intercept_

print('w_norm ', w_norm, 'b_norm ', b_norm)
```

```python
y_pred_sgd = sgd.predict(X_norm)

y_pred = np.dot(X_norm , w_norm) + b_norm

print(y_pred_sgd[:4])
print(y_pred[:4])
```

```python
fig, ax = plt.subplots(1, 4, figsize=(12, 5), sharey=True)
for i in range(len(ax)):
    ax[i].scatter(X_train[:,i], y_train, label='target')
    ax[i].set_xlabel(X_features[i])
    ax[i].scatter(X_train[:, i], y_pred_sgd, label='Predicted')

    
```
