```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# load the dataset
df = pd.read_csv('Housing.csv')
df.head()
```

```python
df.drop(df.iloc[:, 5:], inplace=True, axis=1)
df.head()

X_train = df.iloc[:, 1:]
y_train = df.iloc[:, 0]

X_features = df.columns[2]

print(X_train)
print(y_train)
print(X_features)

```

```python
fig, ax = plt.subplots(1, 4, figsize=(12, 3), sharey=True)
for i in range(len(ax)):
    ax[i].scatter(X_train.iloc[:, i], y_train)
    ax[i].set_xlabel(X_features[i])
```
